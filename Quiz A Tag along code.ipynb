```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/SakyTempress/HDSC-2022-Stage-A/main/FoodBalanceSheets_E_Africa_NOFLAG.csv'
food_data = pd.read_csv(url, error_bad_lines=False, encoding='latin-1')

food_data.head()
```

```python
# What is the total sum of Animal Fat produced in 2014 and 2017 respectively?
food_data.groupby('Item')['Y2014', 'Y2017'].sum()
```

```python
# What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?
food_data.describe()
```

```python
# What is the total number and percentage of missing data in 2016 to 2 decimal places?
food_data.isnull().sum()
```

```python
# Which year had the highest correlation with ‘Element Code’? 
food_data.corr().sum()

```

```python
# What year has the highest sum of Import Quantity?
food_data.groupby('Element')['Y2014', 'Y2015', 'Y2016', 'Y2017'].sum()
```

```python
# What is the total number of the sum of Production in 2014?
food_data.groupby('Element')['Y2014'].sum().sort_values(ascending=False)
```

```python
# Which of these elements had the highest sum in 2018?
food_data.groupby('Element')['Y2015'].sum().sort_values(ascending=False)
```

```python
# Which of these elements had the 3rd lowest sum in 2018?
food_data.groupby('Element')['Y2015'].sum().sort_values(ascending=True)
```

```python
# What is the total Import Quantity in Algeria in 2018?
food_data.groupby(['Area', 'Element'])[ 'Y2018'].sum().head(10)
```
